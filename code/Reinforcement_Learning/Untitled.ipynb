```python
import matplotlib.pyplot as plt
import random
import time

# 그래프 초기 설정
plt.ion()

# 그래프 생성
fig, ax = plt.subplots()

# 초기 데이터
x_data = []
y_data = []

while True:
    # 데이터 갱신
    x = time.time()
    y = random.randint(0, 100)

    x_data.append(x)
    y_data.append(y)

    # 그래프 업데이트
    ax.clear()
    ax.plot(x_data, y_data)

    # 그래프 축 설정 (예: 10초 이내 데이터만 표시)
    ax.set_xlim(x - 10, x)
    ax.set_ylim(min(y_data), max(y_data))

    # 그래프 보여주기
    plt.show()
    plt.pause(10)
```

```python
import random
import numpy as np
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
 
plt.style.use('fivethirtyeight')
 
x_val = []
y_val = []
 
index = count()
 
def animate(i):
    x_val.append(next(index))
    y_val.append(random.randint(0,5))
    plt.cla()
    plt.plot(x_val, y_val)
    
ani = FuncAnimation(plt.gcf(), animate, interval = 1000)
plt.tight_layout()
plt.show()
```

```python
new_real = real[0][10:7813]
new_gru_output_100 = gru_output_100[0][0:7803]

print(new_real.shape)
print(new_gru_output_100.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

# Function to plot the data
def plot_data(real_data, gru_data, start_idx):
    plt.plot(real_data[start_idx:start_idx + 100], label='Real Data', color='blue')
    plt.plot(gru_data[start_idx:start_idx + 100], label='GRU Output', color='red')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.title(f'Data from index {start_idx} to {start_idx + 99}')
    plt.legend()
    plt.show()

# Loop to update the plot every 10 seconds
for i in range(0, min(len(new_real), len(new_gru_output_100)), 100):
    clear_output(wait=True)
    plt.figure(figsize=(10, 5))
    plot_data(new_real, new_gru_output_100, i)
    time.sleep(1)
```
